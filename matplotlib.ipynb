```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
gas= pd.read_csv('gas_prices.csv')
#display contents of the gas
gas
#check NaN vlaues of present
gas.isna().sum()
```

```python
#alter the sizie of the figure
plt.figure(figsize=(8,5))
#add title on top of the plot
plt.title('Gas Prices Over Time (in USD)')
#plot a graph begin usa against the Year
plt.plot(gas.Year, gas['USA'], 'g.-', label='USA')
plt.plot(gas.Year, gas["Canada"],'b.-', label='Canada')
plt.plot(gas.Year,gas['South Korea'],'y.-',label='South Korea')
#add legend to show whose graph curve is
plt.legend()
#skip 3 years
plt.xticks(gas.Year[::3])
#label the x and y 
plt.xlabel('Year')
plt.ylabel('US Dollars')

#show the graph
plt.show()
#save the graph
plt.savefig('graph_one.png',dpi=300)
```

```python
#using the iteration mechanism

#alter the sizie of the figure
plt.figure(figsize=(15, 6))
#add title on top of the plot
plt.title('Gas Prices Over Time (in USD)')
#slabel X and Y
plt.xlabel('Year')
plt.ylabel('Price in USD')
#skip by 2 years
plt.xticks(gas.Year[::3])
for col in gas:
    if col!='Year':
        print(col)
        plt.plot(gas.Year, gas[col],marker='.', label=col)

#add legend to enable labling on the graph        
plt.legend()

#show the graph
plt.show()

#save the graph
plt.savefig('graph_two.png',dpi=300)
```

### Load Fifa Data

```python
#load and read the fifa data 
fifa=pd.read_csv('fifa_data.csv',)
fifa.head()

```

### Histogram

```python
#plotting using the overal skill as basis
plt.hist(fifa.Overall)
plt.show()
```

## PieChart

```python
#getting players with left footed and rieght footed
left_players=fifa.loc[fifa['Preferred Foot']=='Left']
right_players=fifa.loc[fifa['Preferred Foot']=='Right']

#count the number of the left and right footed players
left=left_players.count()[0]
right=right_players.count()[0]

#labels for the pie charts
labels=['left', 'right']
plt.title('Foot Preference of Players')

plt.pie([left,right], labels=labels,shadow=True, autopct='%.3f')
plt.show()

```
